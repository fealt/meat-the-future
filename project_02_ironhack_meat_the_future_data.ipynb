# Data for Meat and Vegetables Consumption

##### Source:
* [Food Balances (2014-), Food and Agriculture Organization of the United Nations](http://www.fao.org/faostat/en/#data/FBS)
* [Food Balances (-2013, old methodology and population), Food and Agriculture Organization of the United Nations](http://www.fao.org/faostat/en/#data/FBSH)

## For Brazil, Germany and World (Total)

##### Details:

* Data can be downloaded as a 'csv' file
* Data had to be downloaded and merged from 2 separate data banks:
    - both from the "Food and Agriculture Organization of the United Natations"
    - but one in an old structure (years before 2013)
    - and another for 2014 onwards

```python
import pandas as pd
import numpy as np
```

```python
# Cleaning and renaming some columns
```

```python
# for 2013 backwards
meat_bovine_2013_minus = pd.read_csv("./assets/originals/meat_bovine_2013_minus.csv", sep=",")
meat_goat_2013_minus = pd.read_csv("./assets/originals/meat_goat_2013_minus.csv", sep=",")
meat_pig_2013_minus = pd.read_csv("./assets/originals/meat_pig_2013_minus.csv", sep=",")
meat_poultry_2013_minus = pd.read_csv("./assets/originals/meat_poultry_2013_minus.csv", sep=",")
meat_other_2013_minus = pd.read_csv("./assets/originals/meat_other_2013_minus.csv", sep=",")

# for 2014 onwards
meat_bovine_2014_plus = pd.read_csv("./assets/originals/meat_bovine_2014_plus.csv", sep=",")
meat_goat_2014_plus = pd.read_csv("./assets/originals/meat_goat_2014_plus.csv", sep=",")
meat_pig_2014_plus = pd.read_csv("./assets/originals/meat_pig_2014_plus.csv", sep=",")
meat_poultry_2014_plus = pd.read_csv("./assets/originals/meat_poultry_2014_plus.csv", sep=",")
meat_other_2014_plus = pd.read_csv("./assets/originals/meat_other_2014_plus.csv", sep=",")
```

```python
# for 2013 backwards
meat_bovine_2013_minus = meat_bovine_2013_minus[
    ["Area Code (FAO)", "Area", "Year", "Value"]
]
meat_goat_2013_minus = meat_goat_2013_minus[["Value"]]
meat_pig_2013_minus = meat_pig_2013_minus[["Value"]]
meat_poultry_2013_minus = meat_poultry_2013_minus[["Value"]]
meat_other_2013_minus = meat_other_2013_minus[["Value"]]

# for 2014 onwards
meat_bovine_2014_plus = meat_bovine_2014_plus[
    ["Area Code (FAO)", "Area", "Year", "Value"]
]
meat_goat_2014_plus = meat_goat_2014_plus[["Value"]]
meat_pig_2014_plus = meat_pig_2014_plus[["Value"]]
meat_poultry_2014_plus = meat_poultry_2014_plus[["Value"]]
meat_other_2014_plus = meat_other_2014_plus[["Value"]]
```

```python
# for 2013 backwards
meat_bovine_2013_minus.columns = [
    "area_code",
    "country",
    "year",
    "meat_bovine_kg_capita",
]
meat_goat_2013_minus.columns = ["meat_goat_kg_capita"]
meat_pig_2013_minus.columns = ["meat_pig_kg_capita"]
meat_poultry_2013_minus.columns = ["meat_poultry_kg_capita"]
meat_other_2013_minus.columns = ["meat_other_kg_capita"]

# for 2014 onwards
meat_bovine_2014_plus.columns = [
    "area_code",
    "country",
    "year",
    "meat_bovine_kg_capita",
]
meat_goat_2014_plus.columns = ["meat_goat_kg_capita"]
meat_pig_2014_plus.columns = ["meat_pig_kg_capita"]
meat_poultry_2014_plus.columns = ["meat_poultry_kg_capita"]
meat_other_2014_plus.columns = ["meat_other_kg_capita"]
```

```python
# for 2013 backwards
meat_consumption_2013_minus = pd.concat(
    [
        meat_bovine_2013_minus,
        meat_goat_2013_minus,
        meat_pig_2013_minus,
        meat_poultry_2013_minus,
        meat_other_2013_minus,
    ],
    axis=1,
)

# for 2014 onwards
meat_consumption_2014_plus = pd.concat(
    [
        meat_bovine_2014_plus,
        meat_goat_2014_plus,
        meat_pig_2014_plus,
        meat_poultry_2014_plus,
        meat_other_2014_plus,
    ],
    axis=1,
)

# all years
meat_consumption = pd.concat(
    [meat_consumption_2013_minus, meat_consumption_2014_plus], axis=0, ignore_index=True
)
meat_consumption
```

```python
meat_consumption["meat_total_kg_capita"] = meat_consumption.iloc[:, 3:8].sum(axis=1)
meat_consumption
```

```python
# for 2013 backwards
veg_2013_minus = pd.read_csv("./assets/originals/veg_2013_minus.csv", sep=",")

# for 2014 onwards
veg_2014_plus = pd.read_csv("./assets/originals/veg_2014_plus.csv", sep=",")
```

```python
# for 2013 backwards
veg_2013_minus = veg_2013_minus[["Area Code (FAO)", "Area", "Year", "Value"]]

# for 2014 onwards
veg_2014_plus = veg_2014_plus[["Area Code (FAO)", "Area", "Year", "Value"]]
```

```python
# for 2013 backwards
veg_2013_minus.columns = [
    "area_code",
    "country",
    "year",
    "vegetables_kg_capita",
]

# for 2014 onwards
veg_2014_plus.columns = [
    "area_code",
    "country",
    "year",
    "vegetables_kg_capita",
]
```

```python
# all years
vegetables_consumption = pd.concat(
    [veg_2013_minus, veg_2014_plus], axis=0, ignore_index=True
)
vegetables_consumption = vegetables_consumption[["vegetables_kg_capita"]]
vegetables_consumption
```

```python
# combining meat and vegetables dataframes

meat_vegetables_consumption = pd.concat(
    [meat_consumption, vegetables_consumption], axis=1
)
meat_vegetables_consumption
```

```python
meat_vegetables = meat_vegetables_consumption.fillna(0)
meat_vegetables['meat_others_kg_capita'] = meat_vegetables['meat_other_kg_capita'] + meat_vegetables['meat_goat_kg_capita']
meat_vegetables.drop(columns=['area_code', 'meat_goat_kg_capita', 'meat_other_kg_capita'], inplace=True)
```

```python
meat_vegetables
```

```python
meat_vegetables['veg_g_capita_day'] = round((meat_vegetables['vegetables_kg_capita'] / 365) * 1000, 2)
meat_vegetables['meat_all_g_capita_day'] = round((meat_vegetables['meat_total_kg_capita'] / 365) * 1000, 2)
meat_vegetables['bov_g_capita_day'] = round((meat_vegetables['meat_bovine_kg_capita'] / 365) * 1000, 2)
meat_vegetables['pig_g_capita_day'] = round((meat_vegetables['meat_pig_kg_capita'] / 365) * 1000, 2)
meat_vegetables['poultry_g_capita_day'] = round((meat_vegetables['meat_poultry_kg_capita'] / 365) * 1000, 2)
meat_vegetables['others_meat_g_capita_day'] = round((meat_vegetables['meat_others_kg_capita'] / 365) * 1000, 2)
```
